```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("vgsalesGlobale.csv")
df
```

```python
df.columns
```

```python
df.drop(['Rank'], axis=1, inplace=True)
```

```python
df
```

```python
#The best-selling game and the amount of sales

df[df["Global_Sales"]==df["Global_Sales"].max()][["Name","Global_Sales"]]
```

```python
#The top 10 best-selling games and the amount of sales

df.sort_values("Global_Sales",ascending=False)[["Name","Global_Sales"]].head(10)
```

```python
#How many different game publishers are there

len(df["Publisher"].unique())
```

```python
#The games that game publishers sell the most and the amount of sales

df.groupby("Publisher").max().sort_values("Global_Sales",ascending=False)[["Name","Global_Sales"]]
```

```python
#Which games have sold more in JP_Sales compared to NA_Sales


df[(df["JP_Sales"]>df["NA_Sales"])][["Name","Genre","JP_Sales","NA_Sales"]]
```

```python
#Games with the word sports in the name


df[df["Name"].str.contains("Sports")]
```

```python
#Application of functions on data

def StrFind(name: str):
    if "sports" in name.lower():
        return True
    else:
        return False

df[df["Name"].apply(StrFind)]
```
